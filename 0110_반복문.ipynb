# 1. While문

```python
# colab에서도 작동되지만 가능하면 .py파일로 작성해서 실행해보세요.
import os

while True:
    userinput = input('>')
    if userinput == 'pwd':
        # print('현재 위치 출력')
        # print('\\'.join(__file__.split('\\')[:-1]))
        print(os.getcwd())
    elif userinput == 'dir' or userinput == 'ls':
        # print('현재 폴더에 폴더와 파일명 출력')
        print(os.listdir(os.getcwd()))
    elif userinput == 'exit':
        print('안녕히가세요.')
        break
```

# 2. Break / continue

* Break = 반복문을 즉시 종료

```python
# 전형적인 for 반복문!
for i in range(10):
    print(i)
# 이 코드는 0부터 4까지 출력하고, i가 5가 되었을 때 반복문을 종료합니다.
```

```python
for i in range(10):
    if i == 5: #만약에 i가 5면 정지시켜라!  -> 0부터 5까지만 출력하고 나머지는 출력하지 않겠다.
        break # break도 코드블록 4칸 띄어서!
    print(i)
```

```python
todo_list = []

def display_todo_list():
    print("현재 할 일 목록:")
    for task in todo_list:
        print("- " + task)
```

```python
def add_task(task):
    todo_list.append(task)
    print(f"'{task}'가 할 일 목록에 추가되었습니다.")

def remove_task(task):
    if task in todo_list:
        todo_list.remove(task)
        print(f"'{task}'가 할 일 목록에서 제거되었습니다.")
    else:
        print(f"할 일 목록에 '{task}'가 없습니다.")
```

```python
while True:
    print("\n[1] 할 일 추가 [2] 할 일 삭제 [3] 목록 표시 [4] 종료")
    choice = input("선택: ")

    if choice == '1':
        task = input("추가할 할 일: ")
        add_task(task)
    elif choice == '2':
        task = input("삭제할 할 일: ")
        remove_task(task)
    elif choice == '3':
        display_todo_list()
    elif choice == '4':
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 입력하세요.")
```

```python
# 할 일 목록을 저장할 리스트
todo_list = []

# 할 일 목록을 표시하는 함수
def display_todo_list():
    print("현재 할 일 목록:")
    for task in todo_list:  # todo_list에 있는 모든 할 일을 순회
        print("- " + task)  # 각 할 일 앞에 '-'를 붙여서 출력

# 할 일을 추가하는 함수
def add_task(task):
    todo_list.append(task)  # 리스트에 새 할 일 추가
    print(f"'{task}'가 할 일 목록에 추가되었습니다.")

# 할 일을 제거하는 함수
def remove_task(task):
    if task in todo_list:  # 할 일이 목록에 있는지 확인
        todo_list.remove(task)  # 목록에서 할 일 제거
        print(f"'{task}'가 할 일 목록에서 제거되었습니다.")
    else:
        print(f"할 일 목록에 '{task}'가 없습니다.")

# 메인 루프
while True:
    print("\n[1] 할 일 추가 [2] 할 일 삭제 [3] 목록 표시 [4] 종료")
    choice = input("선택: ")  # 사용자 입력 받기

    if choice == '1':  # 할 일 추가
        task = input("추가할 할 일: ")
        add_task(task)
    elif choice == '2':  # 할 일 삭제
        task = input("삭제할 할 일: ")
        remove_task(task)
    elif choice == '3':  # 할 일 목록 표시
        display_todo_list()
    elif choice == '4':  # 프로그램 종료
        print("프로그램을 종료합니다.")
        break  # while 루프 종료
    else:
        print("잘못된 선택입니다. 다시 입력하세요.")
```
